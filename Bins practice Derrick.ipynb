```python
# Height is in inches
binned_ht = pd.cut(df['Age'], [-np.inf, 18,30,40,50, np.inf])
binned_ht
```

```python
binned_ht.value_counts(dropna=False)
```

```python
# can add custom labels
df['age_bin'] = pd.cut(
    df['Age'], [-np.inf, 18,30,40,50, np.inf], 
    labels = ["0 to 18'", "18 to 30", "30 to 40'", "40 to 50","Over 50'"]
)
df.drop(['Age'], axis=1, inplace=True)
df['age_bin'].value_counts(dropna=False)
```
